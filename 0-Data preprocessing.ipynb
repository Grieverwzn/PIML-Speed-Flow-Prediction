```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
flow = pd.read_csv('data/flow.csv')
speed = pd.read_csv('data/speed.csv')
flow.shape
```

```python
flow['Timestamp'] = pd.to_datetime(flow['Timestamp'])
flow.set_index('Timestamp', inplace=True)
speed['Timestamp'] = pd.to_datetime(speed['Timestamp'])
speed.set_index('Timestamp', inplace=True)

flow = flow.sort_index()
speed = speed.sort_index()

flow
```

```python
np.random.seed(5)

# Count total days
total_days = (flow.index.max() - flow.index.min()).days + 1

# 30 days of data are randomly selected as the test set
test_days = np.random.choice(total_days, 30, replace=False)

# test set
test_flow = flow.loc[flow.index.map(lambda x: (x - flow.index.min()).days + 1).isin(test_days)]
test_speed = speed.loc[speed.index.map(lambda x: (x - speed.index.min()).days + 1).isin(test_days)]
```

```python
training_flow = flow.loc[~flow.index.map(lambda x: (x - flow.index.min()).days + 1).isin(test_days)]
training_speed = speed.loc[~speed.index.map(lambda x: (x - speed.index.min()).days + 1).isin(test_days)]
```

```python
training_flow.to_csv('data/training_flow.csv')
training_speed.to_csv('data/training_speed.csv')

test_flow.to_csv('data/test_flow.csv')
test_speed.to_csv('data/test_speed.csv')
```

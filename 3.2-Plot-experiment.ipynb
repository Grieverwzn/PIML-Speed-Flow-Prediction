```python
def rmse(y_true, y_pred):

    return np.sqrt(np.mean((y_true - y_pred) ** 2))

def mae(y_true, y_pred):

    return np.mean(np.abs(y_true - y_pred))

def mape(y_true, y_pred):
   
    mask = y_true != 0
    return np.mean(np.abs((y_true - y_pred) / y_true)[mask])
```

```python
gt1_q = q.iloc[:,:150]
mape_q = np.abs(gt1_q.values-q1)/gt1_q.values

mean_q = np.zeros(mape_q.shape[0])
for i in range(mape_q.shape[0]):
    finite_values = mape_q[i, np.isfinite(mape_q[i])]
    mean_q[i] = np.mean(finite_values) if finite_values.size > 0 else 0
    
df_q = pd.DataFrame(mean_q)
df_q.index = gt1_q.index
df_q = df_q.resample('D').mean().dropna()
# df_f
```

```python
gt1_v = v.iloc[:,:150]
mape_v = np.abs(gt1_v.values-v1)/gt1_v.values

mean_v = np.zeros(mape_v.shape[0])
for i in range(mape_v.shape[0]):
    
    finite_values = mape_v[i, np.isfinite(mape_v[i])]
    mean_v[i] = np.mean(finite_values) if finite_values.size > 0 else 0
    
df_v = pd.DataFrame(mean_v)
df_v.index = gt1_v.index
df_v = df_v.resample('D').mean().dropna()
# df_v
```

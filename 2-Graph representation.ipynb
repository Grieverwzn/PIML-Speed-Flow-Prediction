```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from geopy.distance import geodesic
```

```python
sensor = pd.read_csv("data/sensor_infomation.csv")
sensor
```

```python
def calculate_distance(sen1, sen2):
    point1 = (sen1['Latitude'], sen1['Longitude'])
    point2 = (sen2['Latitude'], sen2['Longitude'])
    return geodesic(point1, point2).kilometers
```

```python
# Proximity matrix construction
num_sensor = sensor.shape[0]
proximity_matrix = np.zeros((num_sensor, num_sensor), dtype=np.int32)

for i in range(len(sensor)):
    
    sensor_i = sensor.iloc[i,:] 
    sensor_cal = sensor[(sensor['Fwy']==sensor_i['Fwy'])&(sensor['Dir']==sensor_i['Dir'])&(sensor['ID']!=sensor_i['ID'])]
    
    if len(sensor_cal)<3:
        proximity_matrix[i,sensor_cal.index]=1
    else:
        for j in sensor_cal.index:
            sensor_cal.loc[j,'distance'] = geodesic((sensor_cal.loc[j,'Latitude'],sensor_cal.loc[j,'Longitude']), 
                                                    (sensor_i['Latitude'],sensor_i['Longitude'])).kilometers
        sensor_cal = sensor_cal.sort_values(by='distance', ascending=True)
        proximity_matrix[i,sensor_cal.index[:3]]=1
```

```python
pd.DataFrame(proximity_matrix).to_csv('data/adj_matrix/proximity_matrix.csv',index=False)
```

```python
# Correlation graph of traffic flow
data_flow = pd.read_csv("data/training_flow.csv").iloc[:,1:]

Correlation_matrix_flow = data_flow.corr()
Correlation_matrix_flow = Correlation_matrix_flow.abs()

pd.DataFrame(Correlation_matrix_flow).to_csv('data/adj_matrix/Correlation_matrix_flow.csv',index=False)
print(Correlation_matrix_flow.shape)
Correlation_matrix_flow
```

```python
# Correlation graph of speed
data_speed = pd.read_csv("data/training_speed.csv").iloc[:,1:]

Correlation_matrix_speed = data_speed.corr()
Correlation_matrix_speed = Correlation_matrix_speed.abs()

pd.DataFrame(Correlation_matrix_speed).to_csv('data/adj_matrix/Correlation_matrix_speed.csv',index=False)
print(Correlation_matrix_speed.shape)
Correlation_matrix_speed
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
# K-nearest neighbor graph-flow
X = data_flow.values
X = np.transpose(X)
X = np.nan_to_num(X, nan=0, posinf=0, neginf=0)

# Establish a knn model
knn = NearestNeighbors(n_neighbors=4)
knn.fit(X)
distances, indices = knn.kneighbors(X) 

knn_matrix_flow = np.zeros((len(X), len(X)))

for i in range(len(X)): 
    for j in indices[i]:
        knn_matrix_flow[i, j] = 1
        knn_matrix_flow[j, i] = 1
    
    knn_matrix_flow[i, i] = 0 

pd.DataFrame(knn_matrix_flow).to_csv('data/adj_matrix/knn_matrix_flow.csv',index=False)
```
